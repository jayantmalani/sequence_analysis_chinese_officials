```python
df.columns.values[48]
```

```python
# Career sequences
jobs = df.ix[:,48:]
```

```python
# Each job code, e.g. '2305' needs to be categorized as either 'A,', 'B,' etc.
# We are NOT doing sequence analysis directly on the the 4-digit job codes.
# Those results would not contain sociologically-useful information
# We must also filter on the start/end time columns (job1s, job1e, etc)
jobs
```

```python
# Get rid of the date columns
date_inds = []
for i in range(len(jobs.columns)):
    label = jobs.columns[i]
    if 's' in label or 'e' in label:
        date_inds.append(i)
jobs.drop(jobs.columns[date_inds],axis=1)
#for date_ind in date_inds:
#    jobs.drop(jobs.columns[date_ind],axis=1)
```
